```python
import pandas as pd

# df_pop 이미 읽어온 상태라고 가정
df = df_pop.copy()

# 1) 항목이 '계' 인 행만 사용
df = df[df["항목"] == "계"].copy()

# 2) 자치구 이름 정리
df["district"] = df["동별(1)"]

# 3) 연도 관련 컬럼만 선택
#    - 숫자 또는 "숫자.숫자" 형태
year_cols = [col for col in df.columns if col.replace('.', '').isdigit()]

# 4) melt로 wide → long 변환
df_melt = df.melt(
    id_vars=["district"],
    value_vars=year_cols,
    var_name="year_age",
    value_name="population"
)

# 5) year_age → year와 age_code(연령대 index) 분리
df_melt[["year", "age_code"]] = df_melt["year_age"].str.split(".", expand=True)

# age_code 없는 경우(전체 인구)는 제외
df_melt = df_melt[df_melt["age_code"].notna()].copy()

# 6) 숫자로 변환
df_melt["year"] = df_melt["year"].astype(int)
df_melt["age_code"] = df_melt["age_code"].astype(int)
df_melt["population"] = df_melt["population"].astype(float)

# 7) 우리가 원하는 age_code만 필터링
#    5~9세 → age_code = 2
#    10~14세 → age_code = 3
df_filtered = df_melt[df_melt["age_code"].isin([2, 3])].copy()

# 8) age_code → 나이대 이름으로 매핑
age_map = {
    2: "pop_5_9",
    3: "pop_10_14"
}
df_filtered["age_group"] = df_filtered["age_code"].map(age_map)

# 9) pivot으로 wide화 (district, year별 나이대 컬럼 생성)
df_final = df_filtered.pivot_table(
    index=["district", "year"],
    columns="age_group",
    values="population",
    aggfunc="sum"
).reset_index()

# 10) 자치구 이름 정리
df_final["district"] = df_final["district"].str.strip()

df_final.head()

df_final["pop_children"] = df_final["pop_5_9"] + df_final["pop_10_14"]

df_wide = df_final.pivot_table(
    index="district",
    columns="year",
    values="pop_children",
    aggfunc="sum"
).reset_index()
```

```python
# 자치구 컬럼 생성
df_child["자치구"] = df_child["자치구별(2)"]

# 서울시/소계/합계/NaN 제거
drop_list = ["서울특별시", "합계", "소계", None]
df_child = df_child[~df_child["자치구"].isin(drop_list)]
df_child = df_child[df_child["자치구"].notna()].copy()

# 모든 연도 컬럼 수집
year_raw_cols = [c for c in df_child.columns if c.replace(".", "").isdigit()]

# 연도만 추출 (2006, 2007, ...)
years = sorted(set(int(c.split(".")[0]) for c in year_raw_cols))

# 최종 컬럼 딕셔너리
result = {"자치구": df_child["자치구"]}

for y in years:
    # 해당 연도의 모든 세부 컬럼 찾기
    cols = [c for c in year_raw_cols if c.startswith(str(y))]

    # 이 순서대로라면
    # 0 = 계_시설수
    # 1 = 계_이용자수
    # 2 = 지역아동센터_시설수   ← 필요
    # 3 = 지역아동센터_이용자수 ← 필요
    # 4 = 상담소_시설수
    # 5 = 상담소_이용자수

    cols_sorted = sorted(cols, key=lambda x: float(x))

    # 지역아동센터 시설수 & 이용자수
    fac_col = cols_sorted[2]   # 시설수
    use_col = cols_sorted[3]   # 이용자수

    # 새로운 컬럼명 생성
    result[f"{y}_지역아동센터_시설수"] = pd.to_numeric(df_child[fac_col], errors="coerce")
    result[f"{y}_지역아동센터_이용자수"] = pd.to_numeric(df_child[use_col], errors="coerce")

# 결과 DataFrame 생성
df_child = pd.DataFrame(result)

df_child.head()
```

연도 컬럼 int로 변환

```python
def convert_year_columns(df):
    new_cols = []
    for col in df.columns:
        # 자치구 컬럼은 그대로
        if col == "자치구":
            new_cols.append(col)
        else:
            try:
                # 연도만 숫자로 변환
                new_cols.append(int(col))
            except:
                # 숫자 변환 안되는건 그대로 둠 (필요 없으면 제거 대상)
                new_cols.append(col)
    df.columns = new_cols
    return df
```

Wide-format -> Long format 변환

```python
def to_long(df, value_name):
    """
    Wide-format → Long-format
    df: 자치구 + 연도 wide 데이터프레임
    value_name: long-format의 값 컬럼명
    """
    long_df = df.melt(id_vars=["자치구"], var_name="연도", value_name=value_name)

    # 연도 int 변환
    long_df["연도"] = pd.to_numeric(long_df["연도"], errors="coerce")

    # 자치구 문자열 통일
    long_df["자치구"] = long_df["자치구"].astype(str).str.strip()

    return long_df
```
