```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Train loss (logged)
train_df = pd.read_csv("loss_compare.csv")
train_df["ckpt_num"] = train_df["checkpoint"].str.extract(r"(\d+)").astype(int)

plt.figure()
for ckpt, g in train_df.groupby("ckpt_num"):
    g = g.sort_values("step")
    plt.plot(g["step"], g["loss"], label=f"checkpoint-{ckpt}")
plt.xlabel("step")
plt.ylabel("train loss")
plt.title("Train Loss by Checkpoint")
plt.legend()
plt.tight_layout()
plt.savefig("train_loss_plot.png", dpi=200)
plt.close()

# 2) Eval loss (3 points)
eval_df = pd.read_csv("eval_loss_compare.csv")
eval_df["ckpt_num"] = eval_df["checkpoint"].str.extract(r"(\d+)").astype(int)
eval_df = eval_df.sort_values("ckpt_num")

plt.figure()
plt.plot(eval_df["ckpt_num"], eval_df["eval_loss"], marker="o")
plt.xticks(eval_df["ckpt_num"])
plt.xlabel("checkpoint step")
plt.ylabel("eval_loss")
plt.title("Eval Loss by Checkpoint")
plt.tight_layout()
plt.savefig("eval_loss_plot.png", dpi=200)
plt.close()

print("Saved: train_loss_plot.png, eval_loss_plot.png")
```
